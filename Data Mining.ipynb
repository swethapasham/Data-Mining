```python
x=data1.iloc[:,0:29]
y=data1.iloc[:,30]
```

```python
x
```

```python
y
```

```python
#for splitting the data into training set and test set
#from sklearn import cross_validation, neighbors
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state=0)
smoteX,smoteY = sm.fit_sample(x_train, y_train)
```

```python
#Building Machine learning model
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=100)
model_rf.fit(smoteX,smoteY)
```

```python
#Model Evaluation
y_rf = model_rf.predict(x_test)
from sklearn.metrics import confusion_matrix, classification_report
cm_rf = confusion_matrix(y_test,y_rf)
cr_rf = classification_report(y_test,y_rf)
```

```python
sns.heatmap(cm_rf,annot=True,cbar=False, cmap = 'Set2')
plt.xlabel('True Values'),plt.ylabel('predicted values')
plt.title('Random Forest')
plt.show()
```
